# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

#### 评估从两个方面进行：结构和内容，即整齐度和干净度

```python
original_data.sample(10)
```

#### 数据符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”，具体来看每行是关于某商品的一次交易，每列是交易相关的各个变量，因此不存在结构性问题

### 评估数据干净度

```python
original_data.info()
```

### 从输出结果来看，数据共有541909条观察值，而Description、CustomerID变量存在缺失值。
### 此外，InvoiceDate 的数据类型应为日期，CustomerID 的数据类型应为字符串，应当进行数据格式转换。

### 评估缺失数据

#### 根据条件提取出缺失的Description数值

```python
original_data[original_data["Description"].isnull()]
```

 #### 有1454条交易数据确实Description
 #### 从输出结果来看，这些缺失Description 的交易数据，UnitPrice 都为0 . 为了验证猜想，需要增加筛选条件，看是否存在Description变量缺失且UnitPrice不为0的数据。

```python
original_data[(original_data["Description"].isnull())&(original_data["UnitPrice"] != 0)]
```

#### 输出结果为空证明缺失Description 的交易数据 同时也不具备有效的UnitPrice值

#### Description 表示产品名称，UnitPrice 表示产品单价，都是进行后续商品交易分析的重要变量，如果它们同时缺失/无效，我们认为数据无法提供有效含义，因此这些后续可以被删除

#### CustomerID变量同样存在缺失值，因此也根据条件提取出缺失观察值

```python
original_data[original_data["CustomerID"].isnull()]
```

#### CustomerID 表示客户编号，不是分析畅销商品的必要变量。并且从输出结果来看，有些CustomerID缺失的交易数据仍然有效，因此保留此变量为空的观察值

#### 评估重复数据

#### 根据数据变量的含义来看，虽然nvoiceNo、StockCode、CustomerID 都是唯一标识符，但一次交易可能包含多件商品，因为InvoiceNo可以存在重复，不同交易可以包含同一件商品，因此StockCode可以存在重复，顾客可以进行多次交易或下单多个商品，因此CustomerID也可以存在重复。

#### 针对此数据集，存在重复数据为正常现象，无需评估重复数据。
#### 但还是需要将完全重复的数据进行去重

#### 评估不一致数据

#### 不一致数据可能存在于 Country 变量中，需要查看是否存在多个不同值指代同一国家的情况

```python
original_data["Country"].value_counts()
```

#### 经过检查发现 U.K. 和 UK 以及 United Kingdom 存在重合，USA 和 United States 存在重合，需要合并

#### 评估无效或错误数据


#### 通过DataFrame 的 describe 方法，对数值统计信息进行快速了解。

```python
original_data.describe()
```

#### 从输出结果来看，Quantity 和 UnitPrice 存在负数，会对后续数值分析产生影响。
#### 因此，我们先筛选从出Quantity数值为负数的观察值，进一步评估其含义

```python
original_data[original_data["Quantity"]<0]
```

#### 从筛选结果来看，Quantity 变量为负数的观察值，InvoiceNo似乎均为C开头，表示订单被取消。
#### 但是此猜测不一定准确，为了验证猜想，需要增加筛选条件，看是否存在 Quantity 变量为负数且 InvoiceNo 不为C开头的观察值。

```python
original_data[(original_data["Quantity"]<0)&(original_data["InvoiceNo"].str[0]!="C")]
```

#### 以上猜想被证实错误，因此还存在 Quantity 变量为负数且 InvoiceNo 不为C开头的观察值。
#### 但根据以上输出结果，这些筛选出的观察值的 UnitPrice 观察值均为0，因此增加 UnitPrice 的条件进行验证。

```python
original_data[(original_data["Quantity"]<0)&(original_data["InvoiceNo"].str[0]!="C")&(original_data["UnitPrice"]!=0)]
```

#### 根据输出结果，猜想得到验证，当 Quantity 变量为负数时，观察值满足以下条件之一：
* InvoiceNo 以C开头，表示订单被取消
* UnitPrice 为0，说明单价为0英镑
#### 这些交易数据均不是有效成交数据，不贡献销售，不在后续分析范围内，因此我们将在数据清理步骤，将 Quantity 为负数的观察值删除。
#### 筛选出UnitPrice为负数的观察值。

```python
original_data[original_data["UnitPrice"]<0]
```

#### 从输出结果来看，UnitPrice为负数的观察值都是坏账调账，不属于实际商品交易数据，因此也在数据清理步骤中也将其删除。

## 清理数据

### 根据评估部分所得到的结论，我们需要进行的数据清理包括：
* __把InvoiceDate变量的数据类型转换为日期时间__
* __把CustomerID变量的数据类型转换为字符串__
* __把Description变量缺失的观察值删除__
* __把Country变量中重复的元素进行替换__
* __将完全重复的数据进行去重__
* __把Quantity变量值为负数的观察值删除__
* __把UnitPrice变量值为负数的观察值删除__

#### 为了区分开经过清理的数据和原始的数据，我们创建新的变量cleaned_data，让他为original—_data复制出的副本，我们之后的清理步骤都将被运用在cleaned_data上。

```python
cleaned_data = original_data.copy()
cleaned_data.head(10)
```

#### __把InvoiceDate变量的数据类型转换为日期时间类型：__

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

#### __把CustomerID变量的数据类型转换为字符串:__


```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
```

##### 需要把变量结尾的“.0”删除

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0,-2)
cleaned_data["CustomerID"]
```

#### __把Description变量缺失的观察值删除，并查看删除后该列空缺值个数和：__

```python
cleaned_data = cleaned_data.dropna(subset=["Description"])
cleaned_data["Description"].isnull().sum()
```

#### __把Country变量中重复的元素进行替换,并检查替换后原元素的个数:__

```python
cleaned_data["Country"].replace({"U.K.":"United Kingdom"},inplace=True)
cleaned_data["Country"].replace({"UK":"United Kingdom"},inplace=True)
cleaned_data["Country"].replace({"United States":"USA"},inplace=True)
cleaned_data["Country"].value_counts() 
```

```python
len(cleaned_data[cleaned_data["Country"] =="U.K."])
```

#### __将完全重复的数据进行去重:__

```python
cleaned_data.drop_duplicates(inplace=True)
cleaned_data
```

#### __把Quantity变量值为负数的观察值删除,并检查替换后Quantity变量值为负数的个数:__

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"]>=0]
len(cleaned_data[cleaned_data["Quantity"]<0])
```

#### __把UnitPrice变量值为负数的观察值删除，并检查替换后Quantity变量值为负数的个数：__

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"]>=0]
len(cleaned_data[cleaned_data["UnitPrice"]<0])
```

## 保存清理后的数据

```python
cleaned_data.to_csv("D:\e_commerce_cleaned.csv",index=False)
```
